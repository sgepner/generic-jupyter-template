Just some simple stuff with plots

```python
from sympy import *
import numpy as np
import math

import matplotlib.pyplot as plt
```

```python
y = Symbol('y')
U = Piecewise(
    (-1, y < -1),
    (1, y > 1),
    ( y, True )
)
dU = U.diff(y)
```

```python
fdU= lambdify(y, dU, 'numpy')
fU = lambdify(y, U, 'numpy')
```

```python
Y = np.linspace(-2,2,1000)
plt.plot(Y,fU(Y))
plt.plot(Y,fdU(Y))
plt.grid()
```
